```python
from trajectoryanalytics.databasetable import DatabaseTable
from bokeh.plotting import figure, output_notebook, show
import psycopg2
conn = psycopg2.connect('''dbname=bus host=43.240.97.33 user=ozzytao password=peach''')
table = DatabaseTable('tm_bus_site_his1',['mach_no'],'geom','site_time', conn)
# table.setIDFilter([(101,)])
table.setRandomID()
moves = table.loadMoves()
runs = moves[0].break_into_runs()
```

```python
import ipywidgets as widgets
from ipywidgets import interact, fixed, interactive
from IPython.display import display
import folium
id_list = [x.get_field_value('mach_no') for x in moves]
data = []
for move in moves:
    data.append({'id':move.get_field_value('mach_no'),'raw':move,'runs':move.break_into_runs()})
initial_runs = [x['runs'] for x in data if x['id']==id_list[0]][0]
def show_run(id,r):
    f_map =  folium.Map(location=[34.75,113.65])
    r.folium_plot_points(f_map,color_field='velocity')
    display(f_map)

id_widget = widgets.Dropdown(
    options = id_list,
    value = id_list[0],
    description="mach_no:"
)

def run_options(runs):
    options = {}
    for r in runs:
        options[str(r)]=r
    return options

run_widget = widgets.Dropdown(
    options = run_options(initial_runs),
    description = 'run:'
)
w = interactive(show_run, id=id_widget, r=run_widget)
display(w)
```
